```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
activity = np.array([0, 35, 60, 90, 120, 140, 170, 210, 230, 260, 290])
abs_440 =  np.array([0, 0.1, 0.2, 0.35, 0.49, 0.6, 0.7, 0.85, 0.98, 1.05, 1.15])
plt.plot(activity, abs_440, 'ko')
```

```python
from scipy.stats import linregress

# Dopasowanie
res = linregress(activity, abs_440)

# Wyniki
print(f"y = {res.slope:.3f}x + {res.intercept:.3f}")
print(f"R² = {res.rvalue**2:.3f}")

plt.scatter(activity, abs_440, label='dane')
plt.plot(activity, res.slope*activity + res.intercept, color='red', label='regresja')
plt.legend()
plt.show()
```
